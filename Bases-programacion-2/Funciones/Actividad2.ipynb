```python
#Funcion actualizar inventario
def actualizar_inventario(inventario1, inventario2):
    inventario_actualizado = inventario1.copy()
    
    for producto, cantidad in inventario2.items():#recorrido
        if producto in inventario_actualizado:
            inventario_actualizado[producto] += cantidad#Suma
        else:
            inventario_actualizado[producto] = cantidad
    
    return inventario_actualizado

#Los diccionarios con esta estructura
#inventario1 = {
#    'producto1': cantidad1,
#    'producto2': cantidad2,}
#
#inventario2 = {
#    'producto1': cantidad1,
#    'producto2': cantidad2,}
```

```python
#repaso de matriz 
def ventas_globales(matriz_ventas):
    ventas_por_region = []
    
    for fila in matriz_ventas:
        ventas_region = sum(fila)#pasa por las filas y suma
        ventas_por_region.append(ventas_region)#las agrega a una lista nueva y la imprime
    
    return ventas_por_region

#Estructura de la matriz 
#matriz_ventas = [
#    [ventas_region1_producto1, ventas_region1_producto2, ],
#   [ventas_region2_producto1, ventas_region2_producto2, ], 
#]
```
