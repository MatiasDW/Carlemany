```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos el sistema dinámico discreto
def sistema_dinamico(N_t):
    return 0.95 * N_t + 5

# (b) Verificar si el sistema es lineal y encontrar la solución general
# El sistema es lineal porque es de la forma N_{t+1} = aN_t + b
# Solución general: N_t = C * (0.95)^t + 100

# (c) Encontrar y clasificar los puntos de equilibrio
# Puntos de equilibrio se encuentran cuando N_{t+1} = N_t
# N* = 0.95N* + 5
# 0.05N* = 5
# N* = 100

def punto_equilibrio():
    return 5 / 0.05

# (d) Verificar si hay puntos periódicos
# Necesitamos verificar si existen N_t tales que N_{t+2} = N_t

# (e) Valor a largo plazo de los ingresos del gimnasio
def ingresos_a_largo_plazo(cuota=25):
    N_inf = punto_equilibrio()
    return N_inf * cuota

# Ejecución del script
if __name__ == "__main__":
    N_0 = 50  # Número inicial de socios
    t_max = 100  # Número de meses para simular

    N_t = [N_0]
    for t in range(t_max):
        N_t.append(sistema_dinamico(N_t[-1]))

    # Graficar la evolución del número de socios
    plt.plot(N_t)
    plt.xlabel('Meses')
    plt.ylabel('Número de socios')
    plt.title('Evolución del número de socios en el gimnasio')
    plt.grid(True)
    plt.show()

    print(f"Punto de equilibrio: {punto_equilibrio()}")
    print(f"Ingresos a largo plazo: {ingresos_a_largo_plazo()} euros")
```
